```python
from statsmodels.tsa.arima_model import ARIMA

def plot_URIAGE():
    i=0
    l_index = list(economy1.name)
    for i in range(25):
        one = ''
        one = TU4.iloc[:,i].append(TU3.iloc[:,i]).append(TU2.iloc[:,i]).append(TU1.iloc[:,i])
        one = one.astype(float)
        name = l_index[i]
        
        try:
            arima_model = ARIMA(one, order=(3,1,1)).fit(dist=False)
        except (np.linalg.LinAlgError,ValueError) as err:
            try:
                arima_model = ARIMA(one, order=(3,1,0)).fit(dist=False)
            except (np.linalg.LinAlgError,ValueError) as err:
                try:
                    arima_model = ARIMA(one, order=(3,0,0)).fit(dist=False)
                except (np.linalg.LinAlgError,ValueError) as err:
                    try:
                        arima_model = ARIMA(one, order=(2,1,1)).fit(dist=False)
                    except (np.linalg.LinAlgError,ValueError) as err:
                        try:
                            arima_model = ARIMA(one, order=(2,1,0)).fit(dist=False)
                        except (np.linalg.LinAlgError,ValueError) as err:
                            try:
                                arima_model = ARIMA(one, order=(3,0,0)).fit(dist=False)
                            except ValueError:
                                pass
        predict = arima_model.predict()
        plt.plot(TU4.iloc[:,i].append(TU3.iloc[:,i]).append(TU2.iloc[:,i]).append(TU1.iloc[:,i]))
        plt.plot(predict)
        
        plt.savefig('U-predict-'+name+'.png')
        
        plt.clf()
        
        #resid1 = arima_model.resid
        #fig = plt.figure(figsize=(6, 4), dpi=72)
        #ax = fig.add_subplot(111)
        #ax.plot(resid1,color='b')
        
        #fig.savefig('resid-'+name+'.png')
```

```python
p_num = 3
d_num = 2
q_num = 3
trial_num = p_num * d_num * q_num
use_columns = ["p", "d", "q", "AIC"]
step_cnt = 0

info_df = pd.DataFrame(index=range(trial_num-2), columns=use_columns)

for p in range(p_num):
    for d in range(d_num):
        for q in range(q_num):
            if p == 0 and q == 0:
                continue
            model = ARIMA(one, order=(p, d, q))
            param_list = [p, d, q]
            for name, val in zip(use_columns[:-1], param_list):
                info_df.iloc[step_cnt][name] = val
            try:
                result = model.fit(disp=False)
                # print(result.aic)
                info_df.iloc[step_cnt]["AIC"] = result.aic
            except:
                pass
            step_cnt += 1
            print("Finish trial No. {}, Param: {}".format(step_cnt, param_list))
```
